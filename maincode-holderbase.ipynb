```python
import pandas as pd
import numpy as np
from pandasql import sqldf
from scipy.stats.mstats import winsorize
```

```python
df = pd.read_csv (r'D:\User\Desktop\thesis\data\blockholders\main data\main\Cleaned_Stocks_Holders Final version.csv')     
```

```python
df=df.rename(columns={"name": "firm"})
df['year'] = (df[['jalaliDate']]/10000).astype(int)
df['Market_cap'] =df.apply(lambda row: ((row['shrout']*row['close_price'])/(10**6)),axis=1)
df
```

```python
df.drop(df[df['Percent'] < 5].index, inplace = True)
```

```python
t = df[['Holder', 'Holder_id', 'type', 'firm','jalaliDate','year','Percent','Market_cap']]
t
```

```python
d=t.sort_values(by=['Holder', 'Holder_id','firm','year'])
```

```python
edited=pd.read_excel(r'D:\User\Desktop\thesis\data\blockholders\change_type1.xlsx')
mapingdict = dict(zip(edited.type,edited.new_type))
d['type'] = d.type.map(mapingdict)     
d
```

```python
gg = d.groupby(['Holder', 'Holder_id','type','year'])
def cc(g):
    
    return len(g['firm'].unique())
t = gg.apply(cc).to_frame().rename(columns = {0:"Firm_Number"})
def cc(g):
    return g['Percent'].mean()
t = t.merge(gg.apply(cc).to_frame().rename(columns = {0:"Mean_yearly_totalownership"}),left_index= True,    right_index=True)
t = t.reset_index()

t.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\Holder\Holders_year_unique firm number.xlsx',index=False)
```

```python
d['Mean_yearly_totalownership']=d.groupby(['Holder', 'Holder_id','type','year']).Percent.transform(np.mean)
d['Mean_yearly_firm_ownership']=d.groupby(['Holder', 'Holder_id','type','firm','year']).Percent.transform(np.mean)
d['Mean_yearly_marketcap']=d.groupby(['firm','year']).Market_cap.transform(np.mean)
t = d.drop_duplicates(subset=['Holder','Holder_id','type','year','firm'],keep='last').sort_values(by = ['Holder','year'])
t.rename(columns={"Percent": "End_year_Percent"}, inplace = True)
t.rename(columns={"Market_cap": "End_year_Market_cap"}, inplace = True)
#t['End_year_Market_cap']=winsorize(t['End_year_Market_cap'],limits=[0.01, 0.01])
#t['Mean_yearly_marketcap']=winsorize(t['Mean_yearly_marketcap'],limits=[0.01, 0.01])
t.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\Holder\TSE_HOLDERSS_FIRM_YEAR.xlsx',index=False)
```

```python
dff=t.sort_values(by=['Holder', 'Holder_id','firm','year'])
```

```python
df_sorted=dff.sort_values(by=['Holder','year'])
df_sorted
```

```python
new_df = pd.DataFrame(columns=['Holder','Holder_id','type','Year','Turnover'])
Holder = []
Holder_id = []
type = []
Year = []
Turnover = []
new_df
```

```python
summation = 0
summation_denom = 0
for i in range(len(df_sorted)) : 
    if i == 0 : 
        summation += (df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])
        summation_denom += 0.5 * (df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])
    else : 
          cond = True
          ind = 1
          if df_sorted.iloc[i][5] != df_sorted.iloc[i-1][5] or df_sorted.iloc[i][1] != df_sorted.iloc[i-1][1]  :
            turnover = summation / summation_denom
            summation = 0
            summation_denom = 0
            Holder.append(df_sorted.iloc[i-1][0])
            Holder_id.append(df_sorted.iloc[i-1][1])
            type.append(df_sorted.iloc[i-1][2])
            Year.append(df_sorted.iloc[i-1][5])
            Turnover.append(turnover)
          while cond :
            if  df_sorted.iloc[i][1] == df_sorted.iloc[i-ind][1] :
                if df_sorted.iloc[i][5] == df_sorted.iloc[i-ind][5] + 1 and df_sorted.iloc[i][3] == df_sorted.iloc[i-ind][3] : 
                    summation += (df_sorted.iloc[i][10]) * np.abs(df_sorted.iloc[i][6]- df_sorted.iloc[i-ind][6])
                    summation_denom += 0.5 * (((df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])) + (df_sorted.iloc[i-ind][10] * (df_sorted.iloc[i-ind][6])))
                    cond = False
                else :
                      ind += 1
            else :
                  cond = False
                  summation += (df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])
                  summation_denom += 0.5 * (df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])
    if i == len(df_sorted) - 1 : 
              ind = 1 
              if df_sorted.iloc[i][1] == df_sorted.iloc[i-ind][1] :
                  if df_sorted.iloc[i][5] == df_sorted.iloc[i-ind][5] + 1 and df_sorted.iloc[i][3] == df_sorted.iloc[i-ind][3] : 
                    summation += (df_sorted.iloc[i][10]) * np.abs(df_sorted.iloc[i][6]- df_sorted.iloc[i-ind][6])
                    summation_denom += 0.5 * (((df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])) + (df_sorted.iloc[i-ind][10] * (df_sorted.iloc[i-ind][6])))
                    turnover = summation / summation_denom
                    Holder.append(df_sorted.iloc[i][0])
                    Year.append(df_sorted.iloc[i][5])
                    Turnover.append(turnover)
                  else : 
                        ind += 1
              else :
                    summation += (df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])
                    summation_denom += 0.5 * (df_sorted.iloc[i][10]) * (df_sorted.iloc[i][6])
                    turnover = summation / summation_denom
                    summation = 0
                    summation_denom = 0
                    Holder.append(df_sorted.iloc[i][0])
                    Holder_id.append(df_sorted.iloc[i-1][1])
                    type.append(df_sorted.iloc[i-1][2])
                    Year.append(df_sorted.iloc[i][5])
                    Turnover.append(turnover)
```

```python
new_df['Holder'] = pd.Series(Holder)
new_df['Holder_id'] = pd.Series(Holder_id)
new_df['type'] = pd.Series(type)
new_df['Year'] = pd.Series(Year)
new_df['Turnover'] = pd.Series(Turnover)
new_df.head(20)
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
df = pd.read_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\Holder\Holders_year_unique firm number.xlsx')
dd = pd.read_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\Holder\turnover_measure.xlsx')
q = '''
SELECT dd.Turnover
FROM dd,df
WHERE dd.Holder = df.Holder AND dd.Holder_id = df.Holder_id AND  dd.type = df.type AND dd.year = df.year
'''
names = (pysqldf(q))
list_turnover = list(names['Turnover'])
df['Turnover_measure'] = list_turnover
```

```python
df['Portfo_size'] = '1'
df.loc[df.Firm_Number > 1,'Portfo_size'] = '2'
df.loc[df.Firm_Number > 2,'Portfo_size'] = '3-5'
df.loc[df.Firm_Number > 5,'Portfo_size'] = '>5'

df['Turnover_type'] = 'Long-term'
df.loc[df.Turnover_measure > 0.5,'Turnover_type'] = 'Medium-term2'
df.loc[df.Turnover_measure > 0.1,'Turnover_type'] = 'Medium-term1'
df.loc[df.Turnover_measure  > 1.5,'Turnover_type'] = 'Short-term'

df.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\Holder\Holder_characteristic.xlsx',index=False)
```
