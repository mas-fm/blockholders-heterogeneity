## Amihud_illiquidity

```python
df = pd.read_csv(r'total_stock_price.csv')
df1=df[['Name','Ticker','Market','J-Date','year','Close','Adj Close', 'Volume', 'Value', 'No',]]
df1['return']=df1.groupby(['Ticker'])[['Adj Close']].pct_change()
df1.drop(df1[df1['return'] .isnull()].index, inplace = True)
df1['divide']=(abs(df1['return']) /df1['Value'])*(10**6)
dg=df1.groupby(['Name', 'Ticker','Market','year'])[['divide']].mean().reset_index()
dg.rename(columns={"divide": "Amihud_Illiquidity"}, inplace = True)
dg.replace([np.inf, -np.inf], np.nan, inplace=True)
dg.dropna(subset=["Amihud_Illiquidity"], how="all", inplace=True)
dg.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\firm\Amihud_Illiquidity.xlsx',index=False)
```

```python
pysqldf = lambda q: sqldf(q, globals())
df = pd.read_excel (r'D:\User\Desktop\thesis\data\tarazname\firm_merging data with trazname\final_cashflow.xlsx')
dd = pd.read_excel (r'D:\User\Desktop\thesis\data\tarazname\firm_merging data with trazname\final_balancesheet.xlsx')
dg = pd.read_excel (r'D:\User\Desktop\thesis\data\tarazname\firm_merging data with trazname\final_incomestatement.xlsx')

dg['Sales_Growth']=dg.groupby(['firm'])[['sales']].pct_change()*100

q = '''
SELECT dg.firm ,dg.year ,dg.sales ,dg.net_income, dg.Sales_Growth ,dd.fixed_asset,dd.total_asset ,dd.total_debt ,dd.CEQ 
FROM dg,dd
WHERE dg.firm = dd.firm AND dg.year = dd.year 
'''
names = (pysqldf(q))


q = '''
SELECT names.firm ,names.year ,names.sales ,names.Sales_Growth ,names.net_income ,names.fixed_asset,names.total_asset ,names.total_debt ,names.CEQ ,df.FCF, df.capex 
FROM names,df
WHERE names.firm = df.firm AND names.year = df.year 
'''
names2 = (pysqldf(q))
names2
t3 = names2.drop_duplicates(subset=['firm','year'])


t3.drop(t3[t3['sales'] .isnull()].index, inplace = True)
t3.drop(t3[t3['capex'] .isnull()].index, inplace = True)
t3.drop(t3[t3['net_income'] .isnull()].index, inplace = True)
t3.drop(t3[t3['fixed_asset'] .isnull()].index, inplace = True)
t3.drop(t3[t3['total_asset'] .isnull()].index, inplace = True)
t3.drop(t3[t3['total_debt'] .isnull()].index, inplace = True)
t3.drop(t3[t3['CEQ'] .isnull()].index, inplace = True)
t3.drop(t3[t3['FCF'] .isnull()].index, inplace = True)
t3.drop(t3[t3['Sales_Growth'] .isnull()].index, inplace = True)

t3.rename(columns={"FCF": "F_C_F"}, inplace = True)


tt = pd.DataFrame(t3)
tt['ROA']=tt['net_income']/tt['total_asset']
tt['Firm_Size']=np.log(tt['total_asset'])
tt['Fixed_Assets']=tt['fixed_asset']/tt['total_asset']
tt['Capital_Expenditures']=tt['capex']/tt['total_asset']
tt['Leverage']=tt['total_debt']/tt['total_asset']
tt['FCF']=tt['F_C_F']/tt['total_asset']

t3['ROA']=winsorize(t3['ROA'],limits=[0.01, 0.01])
t3['Sales_Growth']=winsorize(t3['Sales_Growth'],limits=[0.01, 0.01])
t3['Firm_Size']=winsorize(t3['Firm_Size'],limits=[0.01, 0.01])
t3['Fixed_Assets']=winsorize(t3['Fixed_Assets'],limits=[0.01, 0.01])
t3['Capital_Expenditures']=winsorize(t3['Capital_Expenditures'],limits=[0.01, 0.01])
t3['Leverage']=winsorize(t3['Leverage'],limits=[0.01, 0.01])
t3['FCF']=winsorize(t3['FCF'],limits=[0.01, 0.01])


tt.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\firm\statements_variables1.xlsx',index=False)

df = pd.read_excel(r'D:\User\Desktop\thesis\data\blockholders\main data\firm\Amihud_Illiquidity.xlsx')
df=df[['Ticker','year','Amihud_Illiquidity']]
df['Amihud_Illiquidity']=winsorize(df['Amihud_Illiquidity'],limits=[0.01, 0.01])

q = '''
SELECT tt.firm,tt.year,tt.sales,tt.net_income,tt.fixed_asset,tt.total_asset,tt.total_debt,tt.CEQ,tt.FCF,tt.capex,tt.ROA,tt.Sales_Growth,tt.Firm_Size,tt.Fixed_Assets,tt.Capital_Expenditures,tt.Leverage,df.Amihud_Illiquidity
FROM df,tt
WHERE tt.firm = df.Ticker AND tt.year = df.year 
'''
names7 = (pysqldf(q))

df2 = names7.drop_duplicates(subset=['firm','year'])
df2.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\firm\statements_variables2.xlsx',index=False)
```

# Tobins_Q

```python
df2 = pd.read_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\firm\final_merg.xlsx')
df2['Tobins_Q']=(df2['Marketcap']+df2['total_asset']-df2['CEQ'])/df2['total_asset']
df2['Tobins_Q']=winsorize(df2['Tobins_Q'],limits=[0.01, 0.01])
```

# lead variables

```python
df2['Tobins_Q_(t+1)'] =df2.groupby(['firm']).Tobins_Q.shift(-1)
df2['ROA_(t+1)'] =df2.groupby(['firm']).ROA.shift(-1)
df2['FCF_(t+1)'] =df2.groupby(['firm']).FCF.shift(-1)
df2['Tobins_Q_(t+2)'] =df2.groupby(['firm']).Tobins_Q.shift(-2)
df2['ROA_(t+2)'] =df2.groupby(['firm']).ROA.shift(-2)
df2['FCF_(t+2)'] =df2.groupby(['firm']).FCF.shift(-2)
df2['Tobins_Q_(t+3)'] =df2.groupby(['firm']).Tobins_Q.shift(-3)
df2['ROA_(t+3)'] =df2.groupby(['firm']).ROA.shift(-3)
df2['FCF_(t+3)'] =df2.groupby(['firm']).FCF.shift(-3)
df2.to_excel (r'D:\User\Desktop\thesis\data\blockholders\main data\firm\final_merg2.xlsx',index=False)
```
