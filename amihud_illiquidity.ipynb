```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'total_stock_price.csv')
```

```python
df
```

```python
df.columns
```

```python
df1=df[['Name','Ticker','Market','J-Date','year','Close','Adj Close', 'Volume', 'Value', 'No',]]
df1
```

```python
df1['return']=df1.groupby(['Ticker'])[['Adj Close']].pct_change()
df1.drop(df1[df1['return'] .isnull()].index, inplace = True)
df1
```

```python
df1['divide']=(abs(df1['return']) /df1['Value'])*(10**6)
df1
```

```python
dg=df1.groupby(['Name', 'Ticker','Market','year'])[['divide']].mean().reset_index()
dg.rename(columns={"divide": "Amihud_Illiquidity"}, inplace = True)
dg
```
